Outlier : menghapus data yang tidak sesuai target market

```python
# Exercise 1: Identifying and Handling Missing Data
import pandas as pd

# Sample dataset with missing values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', None],
    'Age': [24, 30, None, 22, 35],
    'Salary': [48000, None, 57000, None, 60000]
}
df = pd.DataFrame(data)

# Filling missing values and dropping rows
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Salary'].fillna(df['Salary'].median(), inplace=True)
df.dropna(subset=['Name'], inplace=True)
print('After cleaning:\n', df)
```

```python
# Exercise 2: Standardizing Categorical Data
# Sample dataset with inconsistent categorical values
data = {
    'Product': ['Laptop', 'Laptop', 'Desktop', 'Tablet', 'Tablet'],
    'Category': ['Electronics', 'electronics', 'Electronics', 'Gadgets', 'gadgets']
}
df = pd.DataFrame(data)

# Standardize category values
df['Category'] = df['Category'].str.upper()
print('Standardized Data:\n', df)
```

```python
import pandas as pd

# Dataset sederhana
data = {
    'Age': [25, 30, None, 22, 25, 30, None, 22, 40],
    'City': ['New York', 'new york', 'Los Angeles', 'Los Angeles', 'Chicago', 'CHICAGO', 'new york', 'NEW YORK', 'Chicago']
}
df = pd.DataFrame(data)
print("Original Dataset:")
print(df)
```

```python
# Standarisasi kolom City ke huruf kapital
df['City'] = df['City'].str.upper()

# Menghapus duplikasi baris
df.drop_duplicates(inplace=True)

print("\nDataset setelah standarisasi City dan menghapus duplikasi:")
print(df)
```

```python
import pandas as pd

# Dataset sederhana
data = {
    'City': ['City A', 'City A', 'City A', 'City B', 'City B', 'City B', 'City C', 'City C', 'City C'],
    'Product': ['Product 1', 'Product 2', 'Product 3'] * 3,
    'Sales': [150, 200, 250, 300, 220, 280, 400, 350, 300],
    'Profit': [50, 70, 90, 110, 80, 120, 150, 130, 100]
}
df = pd.DataFrame(data)
print(df)
```

Multi Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np

# Mengelompokkan data dan menghitung rata-rata
grouped = df.groupby(['City', 'Product']).mean().reset_index()
cities = df['City'].unique()
products = df['Product'].unique()

# Menyiapkan data untuk multi-bar chart
x = np.arange(len(products))  # Indeks untuk produk
width = 0.2  # Lebar bar

fig, ax = plt.subplots(figsize=(10, 6))

# Menambahkan bar untuk setiap kota
for i, city in enumerate(cities):
    city_data = grouped[grouped['City'] == city]
    ax.bar(x + i * width, city_data['Sales'], width, label=city)

# Kustomisasi plot
ax.set_xlabel('Products')
ax.set_ylabel('Average Sales')
ax.set_title('Average Sales by Product and City')
ax.set_xticks(x + width)
ax.set_xticklabels(products)
ax.legend(title='Cities')

# Menampilkan plot
plt.tight_layout()
plt.show()
```

Scatter Plot dengan Seaborn

```python
import seaborn as sns

# Scatter plot untuk hubungan Sales dan Profit
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

sns.scatterplot(
    x='Sales', 
    y='Profit', 
    data=df, 
    hue='City', 
    style='Product', 
    palette='viridis', 
    size='Sales', 
    sizes=(50, 200)
)

# Kustomisasi plot
plt.title('Relationship Between Sales and Profit by City and Product', fontsize=14)
plt.xlabel('Sales', fontsize=12)
plt.ylabel('Profit', fontsize=12)
plt.legend(title='City & Product', loc='best')
plt.tight_layout()

# Menampilkan plot
plt.show()
```

Kustomisasi

```python
fig, ax = plt.subplots(figsize=(10, 6))

colors = ['#1f77b4', '#ff7f0e', '#2ca02c']  # Warna untuk tiap kota
for i, city in enumerate(cities):
    city_data = grouped[grouped['City'] == city]
    ax.bar(
        x + i * width, 
        city_data['Sales'], 
        width, 
        label=city, 
        color=colors[i], 
        edgecolor='black'
    )

# Menambahkan grid
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Kustomisasi plot
ax.set_xlabel('Products')
ax.set_ylabel('Average Sales')
ax.set_title('Average Sales by Product and City')
ax.set_xticks(x + width)
ax.set_xticklabels(products)
ax.legend(title='Cities')

plt.tight_layout()
plt.show()
```

# Data Cleaning and Preparation

```python
import pandas as pd

# Membaca file weather_data.csv
df = pd.read_csv('D:/weather_data.csv')

# Mengecek missing values untuk setiap kolom
print("Missing values per column:")
print(df.isnull().sum())

# Mengecek jika ada missing values di seluruh dataset
print("\nApakah ada missing values di dataset?")
print(df.isnull().any().any())

# Menampilkan baris dengan missing values
print("\nBaris dengan missing values:")
print(df[df.isnull().any(axis=1)])
```

```python
# Normalisasi kolom numerik (contoh kolom 'Temperature')
if 'Temperature_C' in df.columns:
    # Mengisi nilai kosong dengan median
    df['Temperature_C'].fillna(df['Temperature_C'].median(), inplace=True)
    
    # Normalisasi menggunakan Min-Max Scaling
    df['Temperature_Normalized'] = (df['Temperature_C'] - df['Temperature_C'].min()) / (df['Temperature_C'].max() - df['Temperature_C'].min())
    print("Dataset setelah normalisasi kolom numerik (Temperature):\n", df.head(), "\n")
```

```python
# Standarisasi kolom kategorikal (contoh kolom 'Condition')
if 'Location' in df.columns:
    df['Location'] = df['Location'].str.upper()  # Standarisasi menjadi huruf kecil

# Menghapus duplikasi
df.drop_duplicates(inplace=True)

print("Dataset setelah standarisasi kolom kategorikal dan menghapus duplikasi:\n", df.head(), "\n")
```

# Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('D:/weather_data.csv')

# Cek apakah kolom yang diperlukan ada
print("Kolom dalam dataset:", df.columns)

# Grouping data berdasarkan lokasi dan menghitung rata-rata
grouped = df.groupby('Location')[['Temperature_C', 'Humidity_pct']].mean()

# Multi-bar chart
locations = grouped.index
x = np.arange(len(locations))
width = 0.35

fig, ax = plt.subplots()
bar1 = ax.bar(x - width/2, grouped['Temperature_C'], width, label='Temperature (C)')
bar2 = ax.bar(x + width/2, grouped['Humidity_pct'], width, label='Humidity (%)')

# Kustomisasi
ax.set_xlabel('Location')
ax.set_ylabel('Values')
ax.set_title('Comparison of Temperature and Humidity by Location')
ax.set_xticks(x)
ax.set_xticklabels(locations)
ax.legend()

# Menampilkan grafik
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('D:/weather_data.csv')

# Cek apakah kolom yang diperlukan ada
print("Kolom dalam dataset:", df.columns)

# Scatter plot dengan kustomisasi layout dan skema warna
if 'Temperature_C' in df.columns and 'Wind_Speed_kmh' in df.columns:
    # Set style dan ukuran layout
    sns.set(style="darkgrid", rc={"figure.figsize": (10, 6)})  # Set style dan ukuran plot
    cmap = sns.color_palette("viridis", as_cmap=True)  # Pilih skema warna
    
    # Membuat scatter plot
    scatter = sns.scatterplot(
        x='Temperature_C', 
        y='Wind_Speed_kmh', 
        data=df, 
        hue='Wind_Speed_kmh', 
        palette=cmap,  # Gunakan skema warna
        size='Wind_Speed_kmh',  # Ukuran titik berdasarkan nilai
        sizes=(50, 200),  # Rentang ukuran titik
        edgecolor='black',  # Warna tepi titik
    )
    
    # Menambahkan kustomisasi lebih lanjut
    plt.title('Relationship Between Temperature and Wind Speed', fontsize=16, fontweight='bold')
    plt.xlabel('Temperature (C)', fontsize=12)
    plt.ylabel('Wind Speed (km/h)', fontsize=12)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    
    # Tambahkan legenda
    scatter.legend(loc='upper left', title='Wind Speed (km/h)', fontsize=10)
    
    # Menampilkan plot
    plt.tight_layout()  # Sesuaikan layout agar tidak terpotong
    plt.show()
else:
    print("Kolom 'Temperature_C' atau 'Wind_Speed_kmh' tidak ditemukan dalam dataset.")
```
